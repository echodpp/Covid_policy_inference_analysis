```python
import pandas as pd
import numpy as np
```

```python
mobility = pd.read_csv('2020_US_Region_Mobility_Report.csv')
```

```python
mobility.head(10)
```

```python
mobility.sample(10)
```

```python
# list of states to filter for
t_states = ['Minnesota', 'Montana', 'Nevada', 'North Carolina', 'Rhode Island', 'Alabama', 'Arizona', 'Florida', 'Georgia', 'Kansas', 'Maine', 'Maryland', 'Mississippi', 'Missouri', 'New Hampshire', 'South Carolina', 'Tennessee', 'Texas', 'Utah', 'Virginia']
c_states = ['Arkansas', 'Iowa' , 'Nebraska', 'North Dakota', 'Oklahoma', 'South Dakota', 'Wyoming']
# filter the mobility dataframe for the specific states
mobility = mobility[mobility['sub_region_1'].isin(t_states) | mobility['sub_region_1'].isin(c_states)]

```

```python
mobility.head(10)
```

```python
# filter the mobility dataframe for the specific states
mobility_state = mobility[mobility['sub_region_2'].isnull()]
```

```python
mobility_state.sample(10)
```

```python
# filter the mobility dataframe for the state county level
mobility_county = mobility[mobility['sub_region_2'].notnull()]
```

```python
mobility_county.sample(10)
```

```python
# analysis of the mobility data
import matplotlib.pyplot as plt
# set the date column as the index
mobility_state.set_index('date', inplace=True)

# create a list of the columns to plot
cols_to_plot = ['retail_and_recreation_percent_change_from_baseline', 'grocery_and_pharmacy_percent_change_from_baseline', 'parks_percent_change_from_baseline', 'transit_stations_percent_change_from_baseline', 'workplaces_percent_change_from_baseline', 'residential_percent_change_from_baseline']

# plot the data for each column
for col in cols_to_plot:
    plt.plot(mobility_state[col], label=col)

# add a legend and labels to the plot
plt.legend()
plt.xlabel('Date')
plt.ylabel('Percent Change from Baseline')
plt.title('Mobility Data for State-Level DataFrame')

# set the x-axis ticks to show only the months and days
plt.xticks(rotation=45, ha='right')
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))

# show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# change the range of the date columns from 2020-03-15 to 2020-04-30
mobility_state_1 = mobility_state[mobility_state.index >= '2020-03-15']
mobility_state_1 = mobility_state_1[mobility_state_1.index <= '2020-04-30']

# filter out the dataframe for c_states and t_states separately
mobility_state_1_c = mobility_state_1[mobility_state_1['sub_region_1'].isin(c_states)]
mobility_state_1_t = mobility_state_1[mobility_state_1['sub_region_1'].isin(t_states)]

# create a list of the mobility columns to plot
mobility_cols = ['retail_and_recreation_percent_change_from_baseline',                  'grocery_and_pharmacy_percent_change_from_baseline',                  'parks_percent_change_from_baseline',                  'transit_stations_percent_change_from_baseline',                  'workplaces_percent_change_from_baseline',                  'residential_percent_change_from_baseline']

# create a subplot for each mobility column
fig, axs = plt.subplots(len(mobility_cols), 1, figsize=(10, 30))

# loop over the mobility columns and plot the control and treatment for each column on the same plot
for i, col in enumerate(mobility_cols):
    axs[i].plot(mobility_state_1_c[col], label='Control')
    axs[i].plot(mobility_state_1_t[col], label='Treatment')

    axs[i].set_xlabel('Date')
    axs[i].set_ylabel('Percent Change from Baseline')
    axs[i].set_title(col)

    axs[i].legend()
    axs[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
